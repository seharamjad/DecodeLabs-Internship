## Task-1: Data Collection & Dataset Understanding

```python
import numpy as np
import pandas as pd 

# Load the Dataset 
df = pd.read_csv('Titanic-Dataset.csv')

# Display the first five rows of the dataset 
print("First five rows of the dataset:")
print(df.head())

# Shape of the dataset (rows, columns)
print("\nSize of the dataset:")
print(df.shape)
```

```python
# Print the Statistical Summary of the dataset 
print("\nStatistical Summary of the Dataset:")
print(df.describe())
```

## Description 

This is the Titanic Dataset. It Contains 891 Rows and 12 Columns. It Represents the Passengers Information from Titanic Ship - including age, gender, ticket class and whether they survived. Key columns include PassengerID, Survived (0=No, 1=Yes), Pclass (ticket class), Name, Sex, Age, SibSp, Parch, Fare and Embarked.
