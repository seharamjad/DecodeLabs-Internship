## Task-2: Data Cleaning & Preprocessing 

##### Goal: Fix missing values, remove duplicates, format data correctly

```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('Titanic-Dataset.csv')

# Step 1: Check missing values 
print('Missing values before cleaning')
print(df.isnull().sum())
```

```python
df.columns.tolist()
```

```python
print(type(df))
print(df.head())
```

```python
# Step 2: Handle the missing values 
print(df.columns)

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])

print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Step 3: Remove duplicates 
print(f'\nDuplicates before cleaning: {df.duplicated().sum()}')

df.drop_duplicates(inplace=True)
print(f'Duplicates after cleaning: {df.drop_duplicates().duplicated().sum()}')
```

```python
# Step 4: Fix Data Types 
# Convert Survived and Pclass to categorical data types 

df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
```

```python
# Step 5: Verify Cleaning 

print('Missing Values after Cleaning')
print(df.isnull().sum())

print('Data Types after Cleaning')
print(df.dtypes)

print('Final Dataset Shape')
print(df.shape)

# Save Cleaned Data
df.to_csv('Titanic-Dataset-cleaned.csv', index=False)
print('Cleaned Dataset saved as Titanic-Dataset-cleaned.csv')
```
