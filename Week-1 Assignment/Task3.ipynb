## Task-3: Exploratory Data Analysis (EDA)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

# Load the Dataset
df = pd.read_csv('Titanic-Dataset-cleaned.csv')

# Basic Statistics 
print('Basic Statistics:')
print(df.describe())

# Survival Count 
print('\nSurvival Count:')
print(df['Survived'].value_counts())
```

```python
# Survival Rate
print('\nSurvival Rate:')
survival_rate = df['Survived'].value_counts(normalize=True) * 100
print(f'Survival Rate: {survival_rate[1]:.2f}%')
print(f'Death Rate: {survival_rate[0]:.2f}%')

# Step 3: Survival by Gender
print('\nSurvival by Gender:')
print(df.groupby('Sex')['Survived'].value_counts())

# Survival by the Class
print('\nSurvival by Ticket Class:')
print(df.groupby('Pclass')['Survived'].value_counts())
```

```python
# Step-5: Data Visualization 
fig,axes = plt.subplots(2,2, figsize=(14,10))
fig.suptitle('Titanic EDA - Key Insights', fontsize=16, fontweight='bold')

# Chart 1 - Survival Count
sns.countplot(x='Survived', data=df, palette='Set2', ax=axes[0,0])
axes[0,0].set_title('Survival Count (0=No, 1=Yes)')
axes[0,0].set_xlabel('Survived')

# Chart 2 - Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1', ax=axes[0,1])
axes[0,1].set_title('Survival by Gender')

# Chart 3 - Survival by Class 
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set3', ax=axes[1,0])
axes[1,0].set_title('Survival by Ticket Class')

# Chart 4 - Age Distribution
axes[1,1].hist(df['Age'], bins=30, color='steelblue', edgecolor='white')
axes[1,1].set_title('Age Distribution of Passengers')
axes[1,1].set_xlabel('Age')
axes[1,1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('EDA_Charts.png', dpi=150)
plt.show()
print('Charts saved as EDA_Charts.png')
```

#### Summary 

1. Only 38% of Passengers survived
2. Women had a much higher Survival Rate than men
3. 1st class passengers survived more than 2nd and 3rd class
4. Most passengers were between 20-40 years old
5. Younger passengers (children) had better survival chances
